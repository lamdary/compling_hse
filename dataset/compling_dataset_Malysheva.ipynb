```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://en.wikipedia.org/wiki/Pig_Latin'
r_wiki = requests.get(url=url)
```

```python
r_wiki
```

```python
r_wiki.text
```

```python
soup_wiki = BeautifulSoup(r_wiki.text, 'html.parser')
soup_wiki
```

```python
pars_wiki = soup_wiki.find_all('p')
pars_wiki
```

```python
texts_wiki = [text.get_text() for text in pars_wiki]
texts_wiki
```

```python
def join_strs(strs):
    result = ''
    for s in strs:
         result += ' ' + s  # а вот и плюс
    return result[1:]
```

```python
joined_wiki = join_strs(texts_wiki)
joined_wiki
```

```python
sp = '!"#$%&()*+,./\:;<=>?@[\]^_`{|}~«»„“...—1234567890'
for item in sp:
  if item in joined_wiki:
    joined_wiki = joined_wiki.replace(item, '')
joined_wiki
```
